Résolution des équations normales

```python
def beta_normal(Xtild, tbar):
    XtX = np.dot(Xtild.T, Xtild)# X^T X
    print(XtX)
    XtX_inv = np.linalg.inv(XtX)# (X^T X)^(-1)
    print(XtX_inv)
    Xtt = np.dot(Xtild.T, tbar)# X^T t
    print(Xtt)
    beta = np.dot(XtX_inv, Xtt)# beta = (X^T X)^(-1) X^T t
    return beta
```
